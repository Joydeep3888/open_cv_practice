```python
#Single Inheritence
import cv2
class image_class:
    
    def __init__(self,path):
        self.path=path
        
    def reading_images_CV(self):
        img=cv2.imread(self.path)
        while True: 
            cv2.imshow('BeyondtheUniverse',img)
            if cv2.waitKey(1) & 0xFF ==27:
                break
        cv2.destroyAllWindows()

class child_of_image_class(image_class): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child(self):
        self.reading_images_CV()
        return('This is the method of child_of_image_class')
```

```python
#mutiple Inheritence
import cv2

class Father:
    
    def __init__(self,path):
        self.path=path
        
    def reading_images_CV(self):
        img=cv2.imread(self.path)
        while True: 
            cv2.imshow('BeyondtheUniverse',img)
            if cv2.waitKey(1) & 0xFF ==27:
                break
        cv2.destroyAllWindows()

class Mother: 
    
    def __init__(self,path):
        pass
    def method_of_child1(self):
        return('This is the method of child_of_image_class')
    
class the_child_class(Father,Mother): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child2(self):
        self.reading_images_CV()
        self.method_of_child1()
        return('Multiple Inheritence: This is the method of the_child_class')
```

```python
x.method_of_child2()
```

```python
#multi level Inheritence
import cv2

class Father:
    
    def __init__(self,path):
        self.path=path
        
    def reading_images_CV(self):
        img=cv2.imread(self.path)
        while True: 
            cv2.imshow('BeyondtheUniverse',img)
            if cv2.waitKey(1) & 0xFF ==27:
                break
        cv2.destroyAllWindows()

class Mother_who_inherites_father_class(Father): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child1(self):
        self.reading_images_CV()
        return('This is the method of child_of_image_class')
    
class the_child_class(Mother_who_inherites_father_class): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child2(self):
        self.reading_images_CV()
        self.method_of_child1()
        return('Multilevel Inheritence: This is the method of the_child_class')
```

```python
#Heirarchical 
import cv2

class Father:
    
    def __init__(self,path):
        self.path=path
        
    def reading_images_CV(self):
        img=cv2.imread(self.path)
        while True: 
            cv2.imshow('BeyondtheUniverse',img)
            if cv2.waitKey(1) & 0xFF ==27:
                break
        cv2.destroyAllWindows()

class Mother_who_inherites_father_class(Father): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child1(self):
        self.reading_images_CV()
        print('Heirarchial Inheritence: This is the method of Mother_who_inherites_father_class')
    
class the_child_class(Father): 
    
    def __init__(self,path):
        self.path=path
    def method_of_child2(self):
        self.reading_images_CV()
        print('Heirarchial Inheritence: This is the method of the_child_class')
```

```python
#Hybrid Inheritance

class School:
    def func1(self):
        print("This function is in school.")
 
 
class Student1(School):
    def func2(self):
        print("This function is in student 1. ")
 
 
class Student2(School):
    def func3(self):
        print("This function is in student 2.")
 
 
class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")
 
 
# Driver's code
object = Student3()
object.func1()
object.func2()
object.func4()
object1=Student2()
object1.func1()
object1.func3()
```

```python

import numpy as np
import cv2

##variables
#True while mouse button down
#false mouse button up

drawing=False
ix,iy=-1,-1

###function##
def draw_rectangle(event,x,y,flags,param):
    global ix,iy,drawing
    
    if event==cv2.EVENT_LBUTTONDOWN:
        drawing=True
        ix,iy=x,y
        
    elif event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            cv2.rectangle(images,(ix,iy),(x,y),(255,0,0),-1)
        
    elif event==cv2.EVENT_LBUTTONUP:
        drawing=False
        cv2.rectangle(images,(ix,iy),(x,y),(255,0,0),-1)

cv2.namedWindow(winname='mydraw')
cv2.setMouseCallback('mydraw',draw_rectangle)

##showingImage###
images=np.zeros((512,512,3),dtype=np.uint8
                #np.uint8
               )
#we can use uint8, int8 or nothing at all, but if we use int16 it will not work
while True:
    cv2.imshow('mydraw',images)
    
    if cv2.waitKey(1) & 0xFF==27:
        break
cv2.destroyAllWindows()

```
