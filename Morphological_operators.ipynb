```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img(): 
    black_img=np.zeros((600,600),dtype=np.float32)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img=black_img,text='ABCDE',org=(100,300),fontFace=font,fontScale=8,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return black_img
```

```python
img=load_img()
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(img)
```

```python
kernal=np.ones((5,5),dtype=np.float32)
```

```python
kernal
```

```python
plt.imshow(kernal)
```

```python
new_img=cv2.erode(img,kernal,iterations=4)
```

```python
display_img(new_img)
```

```python
img=load_img()
white_noise=np.random.randint(low=0,high=2, size=(600,600))
```

```python
white_noise
```

```python
display_img(white_noise)
```

```python
img.max()
```

```python
white_noise=white_noise*255
```

```python
noise_img=white_noise+img
```

```python
display_img(noise_img)
```

```python
#to eliminate the background noise

image=cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernal)
```

```python
display_img(image)
```
